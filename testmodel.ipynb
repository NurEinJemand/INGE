```python
import numpy as np 
import pandas as pd 
import tensorflow as tf
import tensorflow as tf
import os

"""for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        os.path.join(dirname, filename)"""

from tensorflow.keras import layers
from tensorflow.keras import activations
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import model_from_json
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
from tensorflow.keras.layers import BatchNormalization
#from tensorflow.keras.utils import np_utils, to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint

import librosa
from librosa import display

import sklearn
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
from sklearn import model_selection
```

```python
path = 'path/to/file'
```

```python
#from pydub import AudioSegment
#audio = AudioSegment.from_file('../input/aufnahmehappy/Aufnahme.m4a')
aufnahme = np.empty([1, 40])
wav, sr = librosa.load(path)
zts = librosa.feature.mfcc(wav, sr=sr, n_mfcc = 40)
for j in range(40):
        var1 = np.mean(zts[j])
        aufnahme[0][j] = var1
aufnahme1 = preprocessing.normalize(aufnahme)
aufnahme = aufnahme1.reshape(aufnahme.shape[0], aufnahme.shape[1], 1)
```

```python
y_aufnahme = ['4']
from sklearn.preprocessing import label_binarize
y_aufnahme = label_binarize(y_aufnahme, classes=['1','2','3','4','5','6','7','8'])
```

```python
from tensorflow.keras import models
model = models.load_model('../input/emodet070060/EmoDet_10_070060.h5')
model.fit(aufnahme, y_aufnahme)
```

```python
pred_test = model.predict(aufnahme)
komppred = pred_test[0]
prediction = np.argmax(pred_test[0])
print(prediction+1)
print(komppred)
```
